```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WB_WDI_SH_H2O_SMDW_UR_ZS.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# Ensure TIME_PERIOD and OBS_VALUE are numeric
df['TIME_PERIOD'] = pd.to_numeric(df['TIME_PERIOD'], errors='coerce')
df['OBS_VALUE'] = pd.to_numeric(df['OBS_VALUE'], errors='coerce')
```

```python
# Filter for years between 2000 and 2022
df_filtered = df[(df['TIME_PERIOD'] >= 2000) & (df['TIME_PERIOD'] <= 2022)]
```

```python
# Group by year and calculate the global average OBS_VALUE
global_trend = df_filtered.groupby('TIME_PERIOD')['OBS_VALUE'].mean().reset_index()
```

```python
global_trend.head()
```

```python
# Create a line plot to show the trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=global_trend, x='TIME_PERIOD', y='OBS_VALUE', marker='o')
plt.title('Global Average Access Trend (2000–2022)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Average Global Trends in Access to Safe Drinking Water (Global)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter for 2000 and 2022 only
df_filtered = df[df['TIME_PERIOD'].isin([2000, 2022])]
```

```python
# Group by country and year, then compute mean OBS_VALUE
grouped = df_filtered.groupby(['REF_AREA_NAME', 'TIME_PERIOD'])['OBS_VALUE'].mean().unstack()
```

```python
# Drop countries missing data for 2000 or 2022
grouped = grouped.dropna()
```

```python
# Calculate improvement (2022 - 2000)
grouped['Improvement'] = grouped[2022] - grouped[2000]
```

```python
# Sort by highest improvement
top_improvers = grouped.sort_values(by='Improvement', ascending=False)
```

```python
# Show top 10 fastest improving countries
print(top_improvers.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot top 10 fastest improving countries
top10 = top_improvers.head(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=top10, x='Improvement', y='REF_AREA_NAME', color = 'skyblue')
plt.title('Top 10 Fastest Improving Countries (2000–2022)')
plt.xlabel('Improvement people using safely managed drinking water services (urban area)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Filter for countries with no progress or decline
stagnant_or_declining = grouped[grouped['Improvement'] <= 0].sort_values(by='Improvement')
```

```python
# Show results
print("Countries that stagnated or declined (2000–2022):")
print(stagnant_or_declining)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot top 10 fastest improving countries
top10 = stagnant_or_declining.head(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=top10, x='Improvement', y='REF_AREA_NAME', color = 'skyblue')
plt.title('Countries that stagnated or declined (2000–2022):')
plt.xlabel('Improvment people using safely managed drinking water services (urban area)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Cek dulu indikator apa saja yang tersedia
print(df['INDICATOR_NAME'].unique())
```

```python
# Daftar negara ASEAN
asean_countries = ['Indonesia', 'Malaysia', 'Singapore', 'Thailand', 'Vietnam',
    'Philippines', 'Cambodia', 'Lao People\'s Democratic Republic',
    'Myanmar', 'Brunei Darussalam', 'Timor-Leste']
```

```python
filtered_df = df[
    (df['REF_AREA_NAME'].isin(asean_countries)) &
    (df['INDICATOR_NAME'] == 'People using safely managed drinking water services, urban (% of urban population)') &
    (df['TIME_PERIOD'].between(2000, 2022))
].copy()
```

```python
filtered_df['OBS_VALUE'] = pd.to_numeric(filtered_df['OBS_VALUE'], errors='coerce')
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(
    data=filtered_df,
    x='TIME_PERIOD',
    y='OBS_VALUE',
    hue='REF_AREA_NAME',
    marker='o'
)

plt.title('Trends in Access to Safe Drinking Water (Urban) in Southeast Asia (2000–2022)', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Urban Population Percentage(%)')
plt.legend(title='Countries')
plt.grid(True)
plt.tight_layout()
plt.show()
```
